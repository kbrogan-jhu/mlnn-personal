```python
import numpy as np

np.random.seed(7)

from keras.models import Sequential
from keras.layers import Dense
# from keras.optimizers import SGD


import matplotlib.pyplot as plt
%matplotlib inline
```

create the XOR dataset using the code provided

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
plt.scatter(*zip(*X), c=y)
```

Building NN with 2,3,4, and 5 hidden layers, each with 2 neurons. Used this source as my major reference to help understand structure:https://www.atmosera.com/blog/building-neural-networks-with-keras-and-tensorflow/#:~:text=This%20network%20contains%20an%20input,output%20layer%20with%20one%20neuron.

```python
activation1 = 'tanh'
activation2 = 'sigmoid'
activation3 = 'relu'
num_neurons = 2
```

```python
model_HL2 = Sequential(
    [
        Dense(num_neurons, input_dim=2, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(1, activation=activation2),
    ]
)


model_HL3 = Sequential(
    [
        Dense(num_neurons, input_dim=2, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(1, activation=activation2),
    ]
)


model_HL4 = Sequential(
    [
        Dense(num_neurons, input_dim=2, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(1, activation=activation2),
    ]
)


model_HL5 = Sequential(
    [
        Dense(num_neurons, input_dim=2, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(num_neurons, activation=activation1),
        Dense(1, activation=activation2),
    ]
)


models = [model_HL2, model_HL3, model_HL4, model_HL5]
```

```python
num_layers = [2,3,4,5]
scores = []

for model in models:
    
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    model.fit(X, y, batch_size=2, epochs=400) 
    
    score = model.evaluate(X, y) 
    scores.append(score[1])
```

```python
scores
```

###  <font color='purple'>Part 1.1 - output</font>

```python
plt.figure()
plt.xlabel('Number of Layers')
plt.ylabel('Accuracy Score')
plt.plot(num_layers, scores)
plt.title("Two Neurons per Layer")
```

```python
import pandas as pd
```

```python
# Read Data in
Q2X = pd.read_csv("marketing_campaign.csv", sep=';')
pd.set_option('display.max_columns', 50)
Q2X.shape
```

```python
# drop NAs to prepare for NN model
Q2X = Q2X.dropna()
Q2X.shape
```

### <font color='purple'>Data Wrangling to prepare for model input</font>

```python
# data wrangling to prepare for NN model input
drop_cols = ['ID','Dt_Customer', 'Z_Revenue','Z_CostContact','Complain',
             'AcceptedCmp3','AcceptedCmp4','AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2']
transform_cols = ['Education','Marital_Status']
target = ['Response']
```

```python
Q2X = Q2X.drop(drop_cols, axis=1)
```

```python
from sklearn import preprocessing
```

```python
Q2Xt = Q2X.drop(target, axis=1)
Q2target = Q2X[target]

enc = preprocessing.OrdinalEncoder()
```

```python
Q2Xt[transform_cols] = enc.fit_transform(Q2Xt[transform_cols])
```

```python
Q2Xt.describe()
```

```python
Q2Xt.shape, Q2target.shape
```

```python
model = Sequential()
model.add(Dense(26, input_dim=18, activation='tanh'))
model.add(Dense(22, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(12, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
model.fit(Q2Xt, Q2target, epochs=800, batch_size=8)

scores = model.evaluate(Q2Xt, Q2target)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
